```python
# I have a list of numbers and I want to add 5 to each number. how do I do this? 
a = [1, 5, 7, 9]
a
```

```python
b = []   #create emplty list
b.append(a[0]+5)   #do append on empty list
b.append(a[1]+5)
b.append(a[2]+5)
b.append(a[3]+5)
```

```python
b
```

```python
#but this is not a good method to do all the time. for big data we cannot apply it. We can do it with for loop.
a = [1, 5, 7, 9]
b = []   # create an empty list
for number in a:
    b.append(number + 5)   #space is imp at the start here, because it will then come under for loop. It is called identation in python.
b
```

```python
# get square of each number
a = [1, 5, 7, 9]
b = []   # create an empty list
for number in a:
    b.append(number ** 2)   #space is imp at the start here, because it will then come under for loop. It is called identation in python.
b
```

```python
# get sum of each number
a = [1, 5, 7, 9]
SUM = 0   # small letter's sum is key word in green color here, anything in green color in python cannot be get defined to anything.
for number in a:
    SUM = SUM + number  
print(SUM)
```

```python
# get multiplication of each number
a = [1, 5, 7, 9]
mul = 1   # here for multiplication we need to define variable with 1 and not with 0.
for number in a:
    mul = mul * number
print(mul) 
```

```python
# make separate lists for even numbers and odd numbers
a = [1, 4, 10, 15, 18, 23]
even = []
odd = []

for number in a:
    if number % 2 == 0:
        even.append(number)
    elif number % 2 == 1:
        odd.append(number)
print("Even numbers:", even, "Odd numbers:", odd) 
```

```python
# make separate lists for even numbers and odd numbers
a = [1, 4, 10, 15, 18, 23, 23.9, 27.6]
even = []
odd = []
fractions = []

for number in a:
    if number % 2 == 0:
        even.append(number)
    elif number % 2 == 1:
        odd.append(number)
    else: fractions.append(number)      
print("Even numbers:", even, "Odd numbers:", odd, "Fractions numbers:", fractions) 
```

```python
# now optimize the function by removing iteration, because for large numbers too many iteration will be there.
def prime (number):
    if number <= 1:
        return "not a Prime"
    for i in range (2, (number//2) + 1):
        print("trying to divide with", i)
        if number % i == 0:
            return "not a Prime"
    return "Prime"
prime(23)
```

```python
# if we optimize more,because even numbers are not neccessary to check as if number is not be divided by 2,it will not be divided by any other even no.
def prime (number):
    if number <= 1:
        return "not a Prime"
    if number % 2 == 0:
        return "not a Prime"
    for i in range (2, (number//2) + 1):
        if i % 2 != 0:
            print("trying to divide with", i)
        if number % i == 0:
            return "not a Prime"
    return "Prime"
prime(25)
```
